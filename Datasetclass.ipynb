```python
import pandas as pd


excel_file = 'Dataset.xlsx'

# Read specific worksheets into separate DataFrames
Unilag_data = pd.read_excel(excel_file, sheet_name='105. PIVOT DATA', skiprows=4)  # Replace 'Sheet1' with your worksheet name
sales_data = pd.read_excel(excel_file, sheet_name='104. PIVOT TABLE', skiprows=4)  # Replace 'Sheet2' with your worksheet name

# Display the first few rows of each dataset
```

```python
sales_data
```

```python
sales_data_cleaned = sales_data.dropna(axis=1, how='any')

sales_data_cleaned
```

```python
Unilag_data.groupby('GENDER').count()
```

```python
# No of staff in each department using gender
Unilag_data.groupby('DEPARTMENT')['GENDER'].count()
```

```python
# No of staff in each department by gender using lastname
Unilag_data.groupby(['DEPARTMENT', 'GENDER'])['LASTNAME'].count()
```

```python
# No of staff in each department by gender using lastname Pivot Table
Unilag_data.pivot_table('SALARY', index='DEPARTMENT', columns='GENDER')
```

```python
sales_data_cleaned['Profit'] = sales_data_cleaned['Sales'] - sales_data_cleaned['COGS']
sales_data_cleaned
```

```python
sales_data_cleaned.groupby('Product')['Profit'].max()
```
